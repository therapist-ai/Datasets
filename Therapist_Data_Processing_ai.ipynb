```python
df2 = pd.read_csv(io.BytesIO(uploaded['counselchat-data.csv']))
```

```python
df2["Problem"] = df2["questionTitle"] + '-' + df2["questionText"]
```

```python
df3 = pd.DataFrame().assign(Problems=df1['Problem'],Answers=df1['answerText'], topic=df1["topic"])
df4 = pd.DataFrame().assign(Problems=df2['Problem'],Answers=df2['answerText'], topic=df2["topics"])
df4['Answers'] = df4['Answers'].apply(lambda x: BeautifulSoup(x, 'html.parser').get_text())
df = pd.concat([df3,df4])
```
